```python
import pandas as pd
# Suppressing Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_train = df_train.drop('Name',axis = 1)
df_test = df_test.drop('Name',axis = 1)
```

```python
df_train['Sex'] = df_train['Sex'].apply(lambda x: 0 if x == 'male' else 1)
```

```python
df_test['Sex'] = df_test['Sex'].apply(lambda x: 0 if x == 'male' else 1)
```

```python
df_train.head()
```

```python
#Drop ticket
df_train = df_train.drop('Ticket',axis = 1)
df_test = df_test.drop('Ticket',axis = 1)
```

```python
df_train.info()
```

```python
df_train.describe()
```

```python
#Drop Cabin
df_train = df_train.drop('Cabin',axis = 1)
df_test = df_test.drop('Cabin',axis = 1)
```

```python
df_test.describe()
```

```python
df_train['Age'] = df_train['Age'].fillna(38)
```

```python
df_test['Age'] = df_test['Age'].fillna(38)
```

```python
df_train.info()
```

```python
df_temp = df_train.groupby('Embarked').count()
```

```python
df_temp.head()
```

```python
df_train['Embarked'] = df_train['Embarked'].fillna('S')
df_test['Embarked'] = df_test['Embarked'].fillna('S')
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
mean_age = 31.348114
std_age = 13.417662
mean_fare = 32.204208
std_fare = 49.693429
```

```python
df_train['Age'] = ( df_train['Age'] - mean_age ) / std_age
df_test['Age'] = ( df_test['Age'] - mean_age ) / std_age
df_train['Fare'] = ( df_train['Fare'] - mean_fare ) / std_fare
df_test['Fare'] = ( df_test['Fare'] - mean_fare ) / std_fare
```

```python
df_train['Embarked_C'] = df_train['Embarked'].apply(lambda x: 1 if x == 'C' else 0)
df_train['Embarked_S'] = df_train['Embarked'].apply(lambda x: 1 if x == 'S' else 0)
df_test['Embarked_C'] = df_test['Embarked'].apply(lambda x: 1 if x == 'C' else 0)
df_test['Embarked_S'] = df_test['Embarked'].apply(lambda x: 1 if x == 'S' else 0)
```

```python
df_train = df_train.drop('Embarked', axis = 1)
df_test = df_test.drop('Embarked', axis = 1)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_test.info()
```

```python
df_test.describe()
```

```python
df_test['Fare'] = df_test['Fare'].fillna(0.068882)
```

```python
df_test.info()
```

```python
y_train = df_train['Survived']
x_train = df_train.drop('Survived', axis = 1)
```

```python
x_test = df_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
model.fit(x_train, y_train)
```

```python
y_test = model.predict(x_test)
```

```python
df_out = pd.DataFrame(y_test)
df_out.to_csv('submission.csv')
```
